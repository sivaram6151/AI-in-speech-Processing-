# A1. Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal. You may use the below code from librosa as a reference. 

```python
import librosa
import matplotlib.pylab as plt
y, sr = librosa.load("ai_in_sp.wav")
librosa.display.waveshow(y, color='cyan')
plt.show()
```

#  A2. Observe the length & magnitude range of the signal from the plot. Observing the plot, relate to the spoken words and silence between the words.

```python
import numpy as np
mag_min=np.min(y)
mag_max=np.max(y)
print("Signal length:",len(y),"samples")
print("Range of Magnitude:",mag_min,"to",mag_max )
print("Rate of Sampling:",sr, "Hz")
```

# A3. Take a small segment of the signal and play it.

```python
import IPython.display as ipd

y1, sr1 = librosa.load("ai_in_sip_2.wav")

start_time = 1  # this is the start time
end_time = 4    # this is the end time
start_sample = int(start_time * sr1)
end_sample = int(end_time * sr1)

#segmentation 

segment = y1[start_sample:end_sample]
plt.figure(figsize=(15, 5))
librosa.display.waveshow(segment, sr=sr1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
ipd.Audio(segment, rate=sr1)
```

```python
num_segments = 5

# Setting the duration of each segment in seconds
segment_duration = 1

segments = []

for _ in range(num_segments):

    max_start_time = len(y) / sr - segment_duration
    start_time = np.random.uniform(0, max_start_time)

    start_sample = int(start_time * sr)
    end_sample = int((start_time + segment_duration) * sr)
    
    segments.append(y[start_sample:end_sample])

# Visualization
plt.figure(figsize=(15, 5))
for i, segment in enumerate(segments):
    plt.subplot(1, num_segments, i + 1)
    librosa.display.waveshow(segment, sr=sr)
    plt.title(f'Segment {i + 1}')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

for i, segment in enumerate(segments):
    print(f'Playing Segment {i + 1}')
    ipd.display(ipd.Audio(segment, rate=sr))
```
